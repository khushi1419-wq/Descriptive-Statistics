Q7: Variance for Decision-Making (Medium)

Two delivery companies track delivery delays (in minutes).

Company A: variance = 6

Company B: variance = 15

Which company is more consistent, and why?

## Answer:

Variance measures how much the data (delivery delays) spreads out from the average.

Lower variance = more consistency

Higher variance = more fluctuation

Company A: variance = 6

Company B: variance = 15

Since 6 < 15, Company A has less variation in delays.

## Final Answer:

Company A is more consistent because its delivery delays vary less from the average, while Company B shows more unpredictable delays.

Q9: Combining Measures (Hard)

A family records their monthly electricity usage (in kWh):

[400, 420, 390, 450, 410]

Find the mean and standard deviation. What do these values together tell you about the family’s energy use
pattern?

## Answer:

***Step 1: Find Mean***

Mean = Sum of all observations/No of observations

Mean = 400+420+390+450+410/5

Mean = 410.

***Step 2: Find Standard Deviations***

Find the Standard Deviation

find deviations from the mean (414) and square them:

(400 − 414) = −14 = 196

(420 − 414) = 6 = 36

(390 − 414) = −24 = 576

(450 − 414) = 36 = 1296

(410 − 414) = −4 = 16

$\text{Variance} = \frac{\sum (x - \mu)^2}{n}$

Variance = 196 + 36 + 576 + 1296 + 16/5

Variance = 2120/5

Variance = 424

Standard Deviation= $\sqrt{424 }$
≈20.6

Standard Deviation ≈ 20.6 kWh

## Final Answer:

Mean (414 kWh) = The family typically uses about 414 kWh per month

Standard deviation (~20.6 kWh) = Their usage doesn’t vary too much from the average.

The family’s electricity usage is fairly consistent, with only moderate variation each month.
